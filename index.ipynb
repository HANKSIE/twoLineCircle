```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
import numpy as np
import math
```

```python
# init

w1 = 11
w2 = 6
a1 = math.radians(70)
a2 = math.radians(50)
r = 3
```

```python
x = np.linspace(-w1, w2)

L1 = -(w1 + x) / math.tan(a1)
L2 = (x - w2) / math.tan(a2)


crossX = (math.tan(a1) * w2 - math.tan(a2) * w1) / ( math.tan(a1) + math.tan(a2))
crossY = -(w1 + crossX) / math.tan(a1)

AC_line = r / math.sin((a1 + a2) / 2);

AG_line = AC_line * math.cos((math.radians(180) - a1 + a2) / 2)
CG_line = AC_line * math.sin((math.radians(180) - a1 + a2) / 2)

centerX = crossX - AG_line
centerY = crossY + CG_line

AD_line = AB_line = r/math.tan((a1+a2)/2)

BE_line = AB_line * math.sin(math.radians(90)-a1)
EA_line = AB_line * math.cos(math.radians(90)-a1)

L1PX = crossX - EA_line
L1PY = crossY + BE_line

DF_line = AD_line * math.sin(math.radians(90)-a2)
AF_line = AD_line * math.cos(math.radians(90)-a2)

L2PX = crossX + AF_line
L2PY = crossY + DF_line
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(111)
ax.set_aspect('equal', adjustable='box')

ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))

ax.xaxis.set_minor_locator(AutoMinorLocator(1))
ax.yaxis.set_minor_locator(AutoMinorLocator(1))

ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

plt.plot(x, L1, 'red')
plt.plot(x, L2, 'blue')

plt.plot(crossX, crossY, marker='o', markersize=5, color="purple")
plt.plot(L1PX, L1PY, marker='o', markersize=5, color="red")
plt.plot(L2PX, L2PY, marker='o', markersize=5, color="blue")


circle = plt.Circle((centerX, centerY), r, color='grey', fill=False)
ax.add_artist(circle)
plt.plot(centerX, centerY, marker='o', markersize=5, color="green")

plt.xlabel('x')
plt.ylabel('y')

plt.axhline(color='black')
plt.axvline(color='black')

plt.grid(linestyle='--')
plt.show()
```
